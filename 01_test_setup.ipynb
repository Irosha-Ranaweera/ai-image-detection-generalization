### Move 20% of training data into validation folder

```python
import os
import random
import shutil

random.seed(42)

# base_dir = "/content/ddata"
base_dir =base_drive_path + "/ddata"
train_dir = os.path.join(base_dir, "train")
val_dir = os.path.join(base_dir, "val")

split_ratio = 0.2  # 20% of train goes to validation

for cls in ["real", "fake"]:
    train_class_dir = os.path.join(train_dir, cls)
    val_class_dir = os.path.join(val_dir, cls)

    images = [f for f in os.listdir(train_class_dir) if os.path.isfile(os.path.join(train_class_dir, f))]
    random.shuffle(images)

    num_val = int(len(images) * split_ratio)
    val_images = images[:num_val]

    for img in val_images:
        src = os.path.join(train_class_dir, img)
        dst = os.path.join(val_class_dir, img)
        shutil.move(src, dst)

    print(f"Moved {num_val} images from train/{cls} to val/{cls}")
```

```python
import os
import hashlib
import random
from pathlib import Path
from collections import Counter, defaultdict

import numpy as np
import pandas as pd
from PIL import Image, ImageFile, UnidentifiedImageError
import matplotlib.pyplot as plt

ImageFile.LOAD_TRUNCATED_IMAGES = True

data_dir = Path(os.environ["DATA_DIR"])   # expects train/val/test and real/fake folders
valid_exts = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}

records = []
corrupted_files = []
hash_to_paths = defaultdict(list)

def file_md5(path, chunk_size=1024 * 1024):
    md5 = hashlib.md5()
    with open(path, "rb") as f:
        while True:
            chunk = f.read(chunk_size)
            if not chunk:
                break
            md5.update(chunk)
    return md5.hexdigest()

for split in ["train", "val", "test"]:
    for cls in ["real", "fake"]:
        folder = data_dir / split / cls
        if not folder.exists():
            print(f"Missing folder: {folder}")
            continue

        for path in folder.iterdir():
            if not path.is_file():
                continue
            if path.suffix.lower() not in valid_exts:
                continue

            rec = {
                "split": split,
                "class": cls,
                "path": str(path),
                "filename": path.name,
                "ext": path.suffix.lower(),
                "size_bytes": path.stat().st_size,
                "width": None,
                "height": None,
                "mode": None,
                "aspect_ratio": None,
                "corrupted": False,
            }

            try:
                with Image.open(path) as img:
                    img.verify()

                with Image.open(path) as img:
                    img = img.convert("RGB")
                    w, h = img.size
                    rec["width"] = w
                    rec["height"] = h
                    rec["mode"] = "RGB"
                    rec["aspect_ratio"] = round(w / h, 4) if h != 0 else None

                # exact file duplicate check
                file_hash = file_md5(path)
                hash_to_paths[file_hash].append(str(path))

            except (UnidentifiedImageError, OSError, ValueError) as e:
                rec["corrupted"] = True
                corrupted_files.append((str(path), str(e)))

            records.append(rec)

df = pd.DataFrame(records)

print("Total scanned files:", len(df))
print("Corrupted files:", df["corrupted"].sum())

# -----------------------------
# 1) Dataset counts and percentages
# -----------------------------
count_table = (
    df.groupby(["split", "class"])
      .size()
      .unstack(fill_value=0)
      .rename_axis(None, axis=1)
)

count_table["total"] = count_table.sum(axis=1)
count_table["real_%"] = (count_table["real"] / count_table["total"] * 100).round(2)
count_table["fake_%"] = (count_table["fake"] / count_table["total"] * 100).round(2)

overall = df.groupby("class").size().to_dict()
overall_total = len(df)

print("\nCounts by split:")
print(count_table)

print("\nOverall counts:")
for k, v in overall.items():
    print(f"{k}: {v} ({v/overall_total*100:.2f}%)")
print("total:", overall_total)

# -----------------------------
# 2) Basic image statistics
# -----------------------------
clean_df = df[~df["corrupted"]].copy()

print("\nImage resolution stats:")
print(clean_df[["width", "height"]].describe())

print("\nMost common resolutions:")
resolution_counts = clean_df.groupby(["width", "height"]).size().sort_values(ascending=False).head(10)
print(resolution_counts)

print("\nFile format counts:")
print(clean_df["ext"].value_counts())

print("\nAspect ratio stats:")
print(clean_df["aspect_ratio"].describe())

# -----------------------------
# 3) Duplicate files
# -----------------------------
duplicate_groups = [paths for paths in hash_to_paths.values() if len(paths) > 1]
num_duplicate_files = sum(len(g) for g in duplicate_groups)

print("\nExact duplicate groups:", len(duplicate_groups))
print("Total files involved in duplicates:", num_duplicate_files)

if duplicate_groups:
    print("\nExample duplicate group:")
    for p in duplicate_groups[0][:5]:
        print(" -", p)

# -----------------------------
# 4) Estimate RGB mean/std from training set
# -----------------------------
def estimate_mean_std(df_subset, max_images=2000, image_size=(224, 224), seed=42):
    sample_df = df_subset.sample(min(max_images, len(df_subset)), random_state=seed)
    means = []
    stds = []

    for path in sample_df["path"]:
        try:
            with Image.open(path) as img:
                img = img.convert("RGB").resize(image_size)
                arr = np.asarray(img, dtype=np.float32) / 255.0
                means.append(arr.mean(axis=(0, 1)))
                stds.append(arr.std(axis=(0, 1)))
        except Exception:
            continue

    means = np.array(means)
    stds = np.array(stds)
    return means.mean(axis=0), stds.mean(axis=0)

train_df = clean_df[clean_df["split"] == "train"]
mean_rgb, std_rgb = estimate_mean_std(train_df)

print("\nEstimated RGB mean:", np.round(mean_rgb, 4))
print("Estimated RGB std :", np.round(std_rgb, 4))

# -----------------------------
# 5) Save a summary table for report
# -----------------------------
summary_rows = []
for split in ["train", "val", "test"]:
    split_df = clean_df[clean_df["split"] == split]
    total = len(split_df)
    real_n = (split_df["class"] == "real").sum()
    fake_n = (split_df["class"] == "fake").sum()
    summary_rows.append({
        "split": split,
        "real": real_n,
        "fake": fake_n,
        "total": total,
        "real_%": round(real_n / total * 100, 2) if total else 0,
        "fake_%": round(fake_n / total * 100, 2) if total else 0,
    })

summary_df = pd.DataFrame(summary_rows)
summary_df.to_csv("dataset_summary.csv", index=False)
print("\nSaved dataset_summary.csv")

# -----------------------------
# 6) Plot: class distribution by split
# -----------------------------
plot_df = summary_df.set_index("split")[["real", "fake"]]
ax = plot_df.plot(kind="bar", figsize=(7, 4))
ax.set_title("Class Distribution by Split")
ax.set_xlabel("Split")
ax.set_ylabel("Number of Images")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# -----------------------------
# 7) Plot: image width and height distributions
# -----------------------------
fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(clean_df["width"], bins=30, alpha=0.6, label="Width")
ax.hist(clean_df["height"], bins=30, alpha=0.6, label="Height")
ax.set_title("Image Dimension Distribution")
ax.set_xlabel("Pixels")
ax.set_ylabel("Frequency")
ax.legend()
plt.tight_layout()
plt.show()

# -----------------------------
# 8) Plot: aspect ratio distribution
# -----------------------------
plt.figure(figsize=(7, 4))
plt.hist(clean_df["aspect_ratio"].dropna(), bins=30)
plt.title("Aspect Ratio Distribution")
plt.xlabel("Width / Height")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# -----------------------------
# 9) Show sample images
# -----------------------------
def show_random_samples(df_in, split="train", n_per_class=4, seed=42):
    rng = random.Random(seed)
    subset = df_in[(df_in["split"] == split) & (~df_in["corrupted"])]

    fig, axes = plt.subplots(2, n_per_class, figsize=(3 * n_per_class, 6))
    classes = ["real", "fake"]

    for row_idx, cls in enumerate(classes):
        cls_df = subset[subset["class"] == cls]
        chosen = cls_df.sample(min(n_per_class, len(cls_df)), random_state=seed)

        for col_idx, (_, row) in enumerate(chosen.iterrows()):
            ax = axes[row_idx, col_idx]
            with Image.open(row["path"]) as img:
                ax.imshow(img.convert("RGB"))
            ax.set_title(f"{split}/{cls}")
            ax.axis("off")

    plt.tight_layout()
    plt.show()

show_random_samples(clean_df, split="train", n_per_class=4)
```
